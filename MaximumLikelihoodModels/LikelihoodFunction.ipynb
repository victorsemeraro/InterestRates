# <center> <u> Maximum Likelihood Estimation </u> </center> 

```python
import numpy as np 

import seaborn as sns
import matplotlib.pyplot as plt 
```

## <center> Set Seaborn Theme </center>

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

## <center> Fair Coin </center> 

## <center> $ p_{H} = 0.5 $ </center>

## <center> $ P(HH | p_{H} = 0.5) = 0.5^{2} = 0.25 $ </center> 

## <center> Unfair Coin </center> 

## <center> $ p_{H} = 0.3 $ </center> 

## <center> $ P(HH | p_{H} = 0.3) = 0.3^{2} = 0.09 $ </center>

```python
N = 100
p = np.arange(1e-3, 1, 1/N)
```

```python
ax = sns.lineplot(
    x = p,
    y = p**2,
)

plt.show()
```

## <center> Bernoulli Trial </center> 

## <center> $ P(k) = {n \choose k} p^{k}(1 - p)^{n-k}$ </center>

## <center> Binomial Coefficient </center>

## <center> $ {n\choose k} = \frac{n!}{k!(n-k)!} $ </center> 

```python
ax = sns.lineplot(
    x = p,
    y = p**2 * (1 - p),
)
```

```python
ax = sns.lineplot(
    x = p,
    y = np.log(p**2 * (1 - p)),
)
```

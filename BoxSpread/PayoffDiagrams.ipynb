# <center> <u> Box Spread Payoff </u> <center> 

```python
import numpy as np
import pandas as pd 

from numba import jit, prange

import seaborn as sns
import matplotlib.pyplot as plt 
```

## <center> Set Seaborn Theme </center>

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

## <center> Option Price Parameters </center> 

## <center> $ K_{1} < K_{2} $ </center>

```python
N = 10000
K1 = 4000
K2 = 4500
S = np.linspace(3000, 5500, N)
```

## <center> Payoff Functions </center> 

## <center> $ P_{bull} = max(S_{t} - K_{1}, 0) + max(-S_{t} + K_{2}, 0) $ </center> 

```python
long_call = np.array([0 if x < K1  else (x - K1) for x in S])
short_call = np.array([0 if x < K2 else (-x + K2) for x in S])
bull_spread = long_call + short_call
```

```python
fig, axes = plt.subplots(1, 2, layout="constrained")

ax1 = sns.lineplot(
    x = S,
    y = long_call,
    ax = axes[0],
    label = "Payoff",
)

ax1.set(
    title = "Long Call Payoff Function",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

ax2 = sns.lineplot(
    x = S,
    y = short_call,
    ax = axes[1],
    label = "Payoff"
)

ax2.set(
    title = "Short Call Payoff Function",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

plt.show()
```

## <center> Bull Call Spread </center> 

```python
ax = sns.lineplot(
    x = S,
    y = bull_spread, 
    label = "Payoff",
)

ax.set(
    title = "Bull Call Spread Payoff Function",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

plt.show()
```

## <center> Payoff Functions </center> 

## <center> $ P_{bear} = max(K_{2} - S_{t}, 0) + max(-K_{1} + S_{t}, 0) $ </center> 

```python
long_put = np.array([0 if x > K2  else (K2 - x) for x in S])
short_put = np.array([0 if x > K1 else (-K1 + x) for x in S])
bear_spread = long_put + short_put
```

```python
fig, axes = plt.subplots(1, 2, layout="constrained")

ax1 = sns.lineplot(
    x = S,
    y = long_put,
    ax = axes[0],
    label = "Payoff"
)

ax1.set(
    title = "Long Put Payoff Function",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

ax2 = sns.lineplot(
    x = S,
    y = short_put,
    ax = axes[1],
    label = "Payoff"
)

ax2.set(
    title = "Short Put Payoff Function",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

plt.show()
```

## <center> Bear Put Spread </center> 

```python
ax = sns.lineplot(
    x = S,
    y = bear_spread, 
    label = "Payoff",
)

ax.set(
    title = "Bear Put Spread Payoff Function",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

plt.show()
```

## <center> Payoff Functions </center> 

## <center> $ P_{box} = P_{bull} + P_{bear} $ </center> 

## <center> $ P_{box} = max(S_{t} - K_{1}, 0) + max(-S_{t} + K_{2}, 0) + max(K_{2} - S_{t}, 0) + max(-K_{1} + S_{t}, 0) $ </center>

## <center> $ P_{box} = K_{2} - K_{1} $ </center> 

```python
box_spread = bull_spread + bear_spread
```

```python
fig, axes = plt.subplots(1, 2, layout="constrained")

ax1 = sns.lineplot(
    x = S,
    y = bull_spread,
    ax = axes[0],
    label = "Payoff"
)

ax1.set(
    title = "Bull Call Spread",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

ax2 = sns.lineplot(
    x = S,
    y = bear_spread,
    ax = axes[1],
    label = "Payoff"
)

ax2.set(
    title = "Bear Put Spread",
    xlabel = "Stock Price",
    ylabel = "PnL",
)

plt.show()
```

## <center> Box Spread </center> 

```python
ax = sns.lineplot(
    x = S,
    y = box_spread, 
    label = "Payoff",
)

plt.show()
```

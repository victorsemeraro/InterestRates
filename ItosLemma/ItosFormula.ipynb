# <center> <u> Ito's Formula </u> </center>

```python
import numpy as np
from numba import jit 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
sns.set_theme(
    style = "darkgrid",
    palette = "bright",
)

plt.rc('figure', figsize = (10, 5))
```

## <center> Stochastic differential Equation </center>

## <center> $ dX_{t} = \mu_{t} dt + \sigma_{t} dB_{t} $ </center> 

## <center> $ X_{t} = \int_{0}^{t} \mu_{s} ds + \int_{0}^{t} \sigma_{s} dB_{s} $ </center>

## <center> $ E[X_{t}] =  \int_{0}^{t} \mu_{s} ds $ </center> 

## <center> $ Var[X_{t}] = \int_{0}^{t} \sigma_{s}^{2} ds $ </center>

```python
N = 100000
dt = 1 / N
rv = np.random.normal(0, 3, N)
arr = np.zeros(N)

for i in range(1, N):
    arr[i] = arr[i-1] + (rv[i] * np.sqrt(dt))
    
ax = sns.lineplot(arr)

plt.show()
```

## <center> Quadratic Variation </center> 

## <center> $ [X]_{t} = lim_{P \to 0} \sum_{k=1}^{N} (X_{t_{k}} - X_{t_{k-1}})^{2} \to \sigma^{2} $ </center> 

```python
qv = 0

for i in range(1, N):
    qv += (arr[i] - arr[i-1])**2
    
print("Quadratic Variation: ", qv)
```

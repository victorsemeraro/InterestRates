## <center> Three Month Treasury Bill </center> 

## <center> $ P = \frac{FV}{(1 + \frac{r}{4})} $ </center>

```python
@jit(nopython = True)
def compute_three_month_treasury_bill_price(bond_yields):
    """
    
    """
    
    FV = 100
    L = len(bond_yields)
    P = np.copy(bond_yields)
    
    for i in range(L):
        P[i] = 100 / (1 + (bond_yields[i] / 4))
    
    return P
```

```python
bill_prices = compute_three_month_treasury_bill_price(three_month_rate)
```

## <center> Ten Year Note </center> 

## <center> $ P = \sum_{i=1}^{N} \frac{C_{i}}{(1 + \frac{r}{2})^{N}} + \frac{FV}{(1 + \frac{r}{2})^{N}}$ </center>

```python
face_value, maturity, coupon = 100, 10, 0.05
```

```python
[face_value * coupon / 2] * maturity * 2
```

```python
cashflow = [face_value * coupon / 2] * maturity * 2
cashflow[-1] += face_value
print(cashflow[-1])
print(np.sum(cashflow))

plt.bar(np.arange(0, len(cashflow)/2,0.5)+0.5, cashflow, width=0.25)
plt.title(f"cashflow for a {100 * coupon}% {maturity} year bond",fontsize=15)
plt.ylabel("cash flow",fontsize=15)
plt.xlabel("years",fontsize=15)
plt.show()
```

```python
@jit(nopython = True)
def compute_ten_year_treasury_note_price(bond_yields):
    """
    
    """
    
    FV = 100
    N = 2 * 10
    L = len(bond_yields)
    P = np.copy(bond_yields)
    
    for i in range(L):
        
        tmp = np.zeros(N)
        
        for j in range(N-1):
            tmp[j] = (100 * (1 + bond_yields[i] / 2)) / (1 + (bond_yields[i] / 2))**N
            
        tmp[N] = 100 / (1 + (bond_yields[i] / 2))**N
        P[i] = np.sum(tmp) / N
    
    return P
```

```python
note_prices = compute_ten_year_treasury_note_price(ten_year_rate)
```

```python
ax = sns.lineplot(
    bill_prices,
    label = "Bill Prices"
)

plt.show()
```

```python
ax = sns.lineplot(
    note_prices,
    label = "Note Prices",
)

plt.show()
```

## <center> Duration </center>

## <center> $ \frac{\partial}{\partial r} = -\frac{FV}{(1 + r)^{2}} $ </center>

```python
@jit(nopython = True)
def compute_duration(bond_yields, bond_prices):
    """
    
    """
    
    FV = 100
    N = len(bond_yields)
    D = np.copy(bond_yields)
    
    for i in range(N):
        D[i] = (1 / bond_prices[i]) * -FV / (1 + bond_yields[i])**2
    
    return D 
```

## <center> Convexity </center>

## <center> $ \frac{\partial}{\partial r^{2}} = 2 \cdot \frac{FV}{(1 + r)^{3}}$ </center> 

```python
@jit(nopython = True)
def compute_convexity(bond_yields, bond_prices):
    """
    
    """
    
    FV = 100
    N = len(bond_yields)
    D = np.copy(bond_yields)
    
    for i in range(N):
        D[i] = (2 / bond_prices[i]) * (FV / (1 + bond_yields[i])**3)
    
    return D 
```

```python
convexity = compute_convexity(interest_rate, prices)
```

```python
ax = sns.lineplot(
    convexity
)

ax.set(
    title = "Treasury Bond Convexity"
)

plt.show()
```
